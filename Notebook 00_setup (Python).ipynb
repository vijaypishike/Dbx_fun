```python
# 00_setup (managed tables only, no DBFS)

PROJECT = "retailpulse_lite"
DB = PROJECT

# Works in most Free/Serverless setups; if CREATE DATABASE fails, try CREATE SCHEMA.
try:
    spark.sql(f"CREATE DATABASE IF NOT EXISTS {DB}")
except Exception:
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {DB}")

spark.sql(f"USE {DB}")

# Quick sanity
print("Using namespace:", DB)
display(spark.sql("SHOW TABLES"))
```
