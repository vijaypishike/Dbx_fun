```python
%sql
USE retailpulse_lite;

CREATE OR REPLACE TABLE bronze_customers AS
SELECT *, current_timestamp() AS _ingest_ts, 'landing_customers_full' AS _source
FROM landing_customers_full;

CREATE OR REPLACE TABLE bronze_products AS
SELECT *, current_timestamp() AS _ingest_ts, 'landing_products_full' AS _source
FROM landing_products_full;

CREATE OR REPLACE TABLE bronze_orders AS
SELECT *, current_timestamp() AS _ingest_ts, 'landing_orders_full' AS _source
FROM landing_orders_full;

CREATE OR REPLACE TABLE bronze_order_items AS
SELECT *, current_timestamp() AS _ingest_ts, 'landing_order_items_full' AS _source
FROM landing_order_items_full;

-- staging incremental
CREATE OR REPLACE TABLE bronze_customers_incr AS
SELECT *, current_timestamp() AS _ingest_ts, 'landing_customers_incr_20250301' AS _source
FROM landing_customers_incr_20250301;

CREATE OR REPLACE TABLE bronze_orders_incr AS
SELECT *, current_timestamp() AS _ingest_ts, 'landing_orders_incr_20250301' AS _source
FROM landing_orders_incr_20250301;
```
